## Stochastic GD

#### Update the implementation of the Batch GD to be Stochastic and rpeat the above plots and evaluations
#### You can try different learning rate and see its effect.
#### Compare to the Batch GD.
#### Do not forget to shuffle the data.

##### Very Important notes:
- <font color='red'>Save your losses and thetas for each iterations (not each epoch).</font>
- <font color='red'>Plot loss vs. iterations (not epochs). This will allow you to see the difference between SGD (stochastic GD) and batch GD.</font>
- <font color='red'>Convergence check should be performed each epoch not each iterations. This allows you to perform more accurate decision. This is because the noisy process could lead to a deciving decision.</font>
- <font color='red'>Your result may be different from the results shown below due to the randomness of the process. As well as, the random shuffle. The results will be different also for two different run of the algorithm.</font>


```python
def Sto_Gd(X,y,alpha,epoch=1000):
    cost=[]
    theta0s=[]
    theta1s=[]
    y_preds=[]
    cost_epoch=[]
    theta_0,theta_1=0,0
    shuffled_indices = np.random.permutation(len(X)) #return a permutation of the indices
    X = X[shuffled_indices]
    y = y[shuffled_indices]
    for i in range(0,epoch):
        for itr in range(len(X)):
            y_predictd=theta_0+X[itr]*theta_1
            loss = (y_predictd-y[itr])**2
            MSE = loss/2
            d_theta_0=y_predictd-y[itr]
            d_theta_1=(y_predictd-y[itr])*X[itr]
            Gradient_Vector=np.array([[d_theta_0],
                                      [d_theta_1]])
            theta_0 =theta_0-(alpha*d_theta_0)
            theta_1 =theta_1-(alpha*d_theta_1)
            cost.append(MSE)
            theta0s.append(theta_0)
            theta1s.append(theta_1)    
            y_predictd=theta_0+X*theta_1
            y_preds.append(y_predictd)
        cost_epoch.append(cost[-1])
        if i>=1:
            if np.linalg.norm(Gradient_Vector)<0.001 or abs(cost_epoch[-1]-cost_epoch[-2])<.001:
                break
    print("Epochs: ",i+1,'\n')
    print("Grgint victor: ",Gradient_Vector,'\n')
    print("y_predict: ",y_predictd,"\n")
    print("y: ",y)
    print("Accuracy: ",r2_score(y,y_predictd),"\n")
    plt.plot(cost,".b-")
    plt.xlabel("Epoc", fontsize = 15)
    plt.ylabel("Cost",fontsize = 15)
    plt.show()
    plt.plot(theta0s,cost,".b-")
    plt.xlabel("Theta 0", fontsize = 15)
    plt.ylabel("Cost",fontsize = 15)
    plt.show()
    plt.plot(theta1s,cost,".b-")
    plt.xlabel("Theta 1", fontsize = 15)
    plt.ylabel("Cost",fontsize = 15)
    plt.show()
    plt.scatter(X,y)
    plt.plot(X,y_predictd,"r-")
    plt.xlabel("X", fontsize = 15)
    plt.ylabel("Y vs Y_hat",fontsize = 15)
    plt.show()
    plt.plot(theta0s,theta1s,".b-")
    plt.xlabel("Theta 0", fontsize = 15)
    plt.ylabel("Theta 1",fontsize = 15)
    plt.show()
    
    plt.scatter(X,y)
    for it in y_preds:
        plt.plot(X,it,'r')
        
        
```

##### The following results uses alpha = 0.0005 and max. iterations = 300.

```python
Sto_Gd(X,y,.0005,300)
```

##### The following results uses alpha = 0.0001 and max. iterations = 300.

```python
Sto_Gd(X,y,.0001,300)
```

##### The following results uses alpha = 0.00007 and max. iterations = 300.

```python
Sto_Gd(X,y,0.00007,300)
```

## Mini-Batch GD

#### Update the implementation of the Batch GD to be Mini-Batch and repeat the above plots and evaluations
#### Change batch size and see the effect
#### Do not forget to shuffle the data

##### Very Important notes:
- <font color='red'>Save your losses and thetas for each iterations (not each epoch).</font>
- <font color='red'>Plot loss vs. iterations (not epochs). This will allow you to see the difference between SGD (stochastic GD) and batch GD.</font>
- <font color='red'>Convergence check should be performed each epoch not each iterations. This allows you to perform more accurate decision. This is because the noisy process could lead to a deciving decision.</font>
-<font color='red'>Your result may be different from the results shown below due to the randomness of the process. As well as, the random shuffle. The results will be different also for two different run of the algorithm.</font>

```python
def mini_Gd(X,y,alpha,mini,epoch=1000):
    cost=[]
    theta0s=[]
    theta1s=[]
    y_preds=[]
    cost_epoch=[]
    theta_0,theta_1=0,0
    shuffled_indices = np.random.permutation(len(X)) #return a permutation of the indices
    X = X[shuffled_indices]
    y = y[shuffled_indices]
    for i in range(0,epoch):
        for itr in range(0,len(X),mini):
            y_predictd=theta_0+X[itr:itr+mini].dot(theta_1)
            loss = np.square(y_predictd-y[itr:itr+mini])
            MSE = np.sum(loss)/(2*len(X[itr:itr+mini]))
            d_theta_0=(1/len(X[itr:itr+mini]))*np.sum((y_predictd-y[itr:itr+mini]))
            d_theta_1=(1/len(X[itr:itr+mini]))*np.sum(np.dot((y_predictd-y[itr:itr+mini]),X[itr:itr+mini]))
            Gradient_Vector=np.array([[d_theta_0],
                                      [d_theta_1]])
            theta_0 =theta_0-(alpha*d_theta_0)
            theta_1 =theta_1-(alpha*d_theta_1)
            cost.append(MSE)
            theta0s.append(theta_0)
            theta1s.append(theta_1)    
            y_preds.append(y_predictd)
        cost_epoch.append(MSE)
        if i>=1:
            if np.linalg.norm(Gradient_Vector)<0.001 or abs(cost_epoch[-1]-cost_epoch[-2])<.001:
                y_predictd=theta_0+X.dot(theta_1) 
                break
    print("Epochs: ",i+1,'\n')
    print("Grgint victor: ",Gradient_Vector,'\n')
    print("y_predict: ",y_predictd,"\n")
    print("y: ",y)
    print("Accuracy: ",r2_score(y,y_predictd),"\n")
    plt.plot(cost,".b-")
    plt.xlabel("Epoc", fontsize = 15)
    plt.ylabel("Cost",fontsize = 15)
    plt.show()
    plt.plot(theta0s,cost,".b-")
    plt.xlabel("Theta 0", fontsize = 15)
    plt.ylabel("Cost",fontsize = 15)
    plt.show()
    plt.plot(theta1s,cost,".b-")
    plt.xlabel("Theta 1", fontsize = 15)
    plt.ylabel("Cost",fontsize = 15)
    plt.show()
    plt.scatter(X,y)
    plt.plot(X,y_predictd,"r-")
    plt.xlabel("X", fontsize = 15)
    plt.ylabel("Y vs Y_hat",fontsize = 15)
    plt.show()
    plt.plot(theta0s,theta1s,".b-")
    plt.xlabel("Theta 0", fontsize = 15)
    plt.ylabel("Theta 1",fontsize = 15)
    plt.show()
          
        
```

##### The following results uses batch size=5, alpha = 0.0005 and max. iterations = 300.

```python
mini_Gd(X,y,.0005,5,300)
```
